```python
import urllib.request, urllib.error, urllib.parse
from bs4 import BeautifulSoup
import pandas as pd
import spacy
from spacy import displacy
import en_core_web_lg

nlp = spacy.load('en_core_web_lg')
import_file = open("Names.txt", "w")

with open("Names.txt", "a") as file1:
    for searchresult in search_results["webPages"]["value"]:
        url = searchresult["url"]
        file1.write(url)#print(url)
        try:  
            response = urllib.request.urlopen(url)
            for line in response:
                line = line.decode('utf-8')
                soup = BeautifulSoup(line)
                for script in soup(["script", "style"]):
                    script.decompose()
                #print(soup.text)
                doc = nlp(soup.text)
                for ent in doc.ents:
                    if(ent.label_ == 'PERSON'):
                        file1.write(ent.text+"\n")#print(ent.text)
                        print("Writing to file...")
        except:
            file1.write(ent.text+"\n")#print("there was an error")
            print("Error Reported to file...")
```
